```python
import matplotlib.pyplot as plt
import mpld3
import numpy as np
import pandas as pd
```

```python
n = 1000
idx = np.arange(n)
value = np.random.rand(n) + 5
datetime = pd.Timestamp.now() + pd.to_timedelta(idx, unit='s')
data = pd.DataFrame(data=zip(datetime, value), index=idx, columns=['datetime', 'value'])
data.sample(5).sort_index()
```

```python
mpld3.enable_notebook()
plt.figure(figsize=(12, 9))
plt.plot(data['datetime'], data['value'])
[plt.plot([x, x], [0, 10], color='r') for x in data['datetime'].sample(3)]
plt.ylim(0, 10)
plt.show()
plt.close()
mpld3.disable_notebook()
```
